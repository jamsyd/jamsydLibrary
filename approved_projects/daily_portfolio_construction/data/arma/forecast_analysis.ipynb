```python
import os
import glob

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
merge_lst = []

path      = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
for f in csv_files:
    if "_forecasts" in f:  
        merge_lst.append(pd.read_csv(f) )

merge_df = pd.concat(merge_lst,axis=0).reset_index()
```

```python
merge_df
```

```python
merge_df.to_csv('forecast_df.csv')
```

# Forecasts

```python
forecast_df = merge_df.copy()
```

```python
products  = ["XBI","XLB","XLC","XLE","XLF","XLI","XLK","XLP","XLU","XLV","XLU","XLY","XME","XRT","XTL","XTN"]
```

```python
for product in products:
    print(product)

    plt.figure(figsize=(45,15))
    plt.plot(forecast_df[forecast_df['product_name'] == product]['Close'])
    plt.plot(forecast_df[forecast_df['product_name'] == product]['pointForecast'])

    plt.show()
```

# Begin Analysis

```python
products  = ["XBI","XLB","XLC","XLE","XLF","XLI","XLK","XLP","XLU","XLV","XLU","XLY","XME","XRT","XTL","XTN"]
```

```python
forecast_df['forecast_error'] = forecast_df['Close'] - forecast_df['pointForecast']
```

```python
for product in products:
    print(product)

    plt.figure(figsize=(45,15))
    plot_df = forecast_df[forecast_df['product_name'] == product]
    i = 1
    while i < 6:

        plt.hist(forecast_df[forecast_df['forecastday'] == i]['forecast_error'],bins=100)
        print(i,np.mean(forecast_df[forecast_df['forecastday'] == i]['forecast_error']))
        print(i,np.std(forecast_df[forecast_df['forecastday'] == i]['forecast_error']))
        i+=1
    plt.legend(["1","2","3","4","5",])



    plt.show()
```
