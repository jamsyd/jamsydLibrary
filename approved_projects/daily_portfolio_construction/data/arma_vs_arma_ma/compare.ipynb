```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
arma_ma_df = pd.read_csv(r'arma_ma_pnl.csv',parse_dates=True)
arma_df = pd.read_csv(r'arma_pnl.csv',parse_dates=True)
```

```python
products  = ["XBI","XLB","XLC","XLE","XLF","XLI","XLK","XLP","XLU","XLV","XLU","XLY","XME","XRT","XTL","XTN"]
```

# PnL Analytics

```python
better = 0
worse = 0

for product in products:

    print(product,"arma_ma")
    print(np.std(arma_ma_df[arma_ma_df['product_name'] == product]['pnl']))
    print(np.mean(arma_ma_df[arma_ma_df['product_name'] == product]['pnl']))
    print(" ")

    print(product,"arma")
    print(np.std(arma_df[arma_df['product_name'] == product]['pnl']))
    print(np.mean(arma_df[arma_df['product_name'] == product]['pnl']))
    print(" ")

    print("scale mean")

    if np.mean(arma_df[arma_df['product_name'] == product]['pnl'])*(np.std(arma_ma_df[arma_df['product_name'] == product]['pnl'])/np.std(arma_ma_df[arma_ma_df['product_name'] == product]['pnl'])) > np.mean(arma_df[arma_df['product_name'] == product]['pnl']):
        print("Better Performance")
        better+=1
    else:
        print("Worse")
        worse+=1
    print(" ")

print(better, worse, better/(better+worse))
```

# Plotting 

```python
for product in products:

    print(product)

    plt.figure(figsize=(45,15))
    plt.plot(arma_ma_df[arma_ma_df['product_name'] == product]['pnl'].cumsum())
    plt.plot(arma_df[arma_df['product_name'] == product]['pnl'].cumsum())

    plt.legend(['arma_ma','arma'])
    plt.show()
```
