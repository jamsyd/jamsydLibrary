```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from statsmodels.tsa.arima_model import ARMA

from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
spy_df = pd.read_csv(r'C:\Users\James Stanley\Documents\GitHub\backtest_utilities\data\daily\stocks\BATS_SPY, 1D.csv',parse_dates=True,index_col='time')
xlu_df = pd.read_csv(r'C:\Users\James Stanley\Documents\GitHub\backtest_utilities\data\daily\stocks\BATS_XLU, 1D.csv',parse_dates=True,index_col='time')
xli_df = pd.read_csv(r'C:\Users\James Stanley\Documents\GitHub\backtest_utilities\data\daily\stocks\BATS_XLI, 1D.csv',parse_dates=True,index_col='time')
```

```python
plt.figure(figsize=(45,15))

plt.plot(spy_df['close'],color='blue')
plt.twinx()
plt.plot(xlu_df['close'],color='red')
plt.twinx()
plt.plot(xli_df['close'],color='green')
plt.show()

plt.figure(figsize=(45,15))

plt.plot(xlu_df['close']/spy_df['close'],color='blue')
plt.plot(xli_df['close']/spy_df['close'],color='orange')

plt.show()
```

```python
rel_df = xli_df['close']/spy_df['close']
```

```python
log_diff = np.log(rel_df).diff(1).dropna(axis=0)
```

```python
plt.figure(figsize=(45,15))

plt.hist(log_diff,bins=100)

plt.show()
```

```python
plot_acf(log_diff, alpha=1, lags=5).savefig("acf.jpg")
```

```python
plot_pacf(log_diff, alpha=1, lags=5).savefig("pacf.jpg")
```
