```python
ngs_df = ngs_df[ngs_df['Source']=='EIA-912']
```

```python
for col in cols:
    plt.figure(figsize=(45,15))
    plt.title(col)
    plt.plot(ngs_df[col])
    plt.show()
```

```python
from statsmodels.tsa.arima_model import ARMA

from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
for col in cols:
    seasonal_result = seasonal_decompose(ngs_df[col], model="additive",period=52)
    seasonal = seasonal_result.seasonal
    plt.figure(figsize=(45,15))
    plt.title(col)
    plt.plot(seasonal)
    plt.show()
```

```python
for col in cols:
    seasonal_result = seasonal_decompose(ngs_df[col], model="additive",period=52)
    residual = seasonal_result.resid
    plt.figure(figsize=(45,15))
    plt.title(col)
    plt.plot(residual)
    plt.show()
```

```python
for col in cols:

    seasonal_result = seasonal_decompose(ngs_df[col], model="additive",period=52)
    residual = seasonal_result.resid
    diff = residual.reset_index().dropna(axis=0)

    plot_acf(diff['resid'], alpha=1, lags=5).savefig(f"{col}.jpg")
```

```python
for col in cols:

    seasonal_result = seasonal_decompose(ngs_df[col], model="additive",period=52)
    residual = seasonal_result.resid
    diff = residual.reset_index().dropna(axis=0)

    plot_pacf(diff['resid'], alpha=1, lags=5).savefig(f"{col}.jpg")
```
